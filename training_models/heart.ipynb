```python
import numpy as np #used for mathematical calculations
import pandas as pd #used for data visualization
from sklearn.model_selection import train_test_split #used to use the ml algorithms drectly
from sklearn import svm #source vector machine
from sklearn.metrics import accuracy_score,precision_score,recall_score
```

```python
heart_dataset = pd.read_csv("../datasets/heart.csv")
```

```python
#now separate the independent as X and dependent as Y
X = heart_dataset.drop(columns=['target'],axis=1)
Y=heart_dataset['target']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=45)
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
X_test.shape
```

```python
Y_test.shape
```

```python
model = svm.SVC(kernel='linear')
```

```python
model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
print(accuracy_score(Y_test,Y_pred))
print(precision_score(Y_test,Y_pred))
print(recall_score(Y_test,Y_pred))
```

```python
import pickle
filename = "heart_model.sav"
pickle.dump(model,open(filename,'wb'))
```
